```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import scipy.stats as st
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
#2 Загрузка и чтение файла
df=pd.read_csv('home_price.csv')
print(df.info())
```

```python
print(df.head(5))
```

```python
#3 Предварительная обработка и очистка данных
print(df.isnull().sum())#посмотрим кол-во пустых значений
print(df.duplicated().sum())#посмотрим кол-во дубликатов
df=df.drop_duplicates() #удалим дубликат
```

```python
#удалим пропущенные значения в колонке floors_total,т.к.их немного,а в колонке is_apartment на заменим 'False'

df=df.dropna(subset='floors_total')
df['is_apartment']=df['is_apartment'].fillna('False')

#преобразуем колонку is_apartment в тип bool
df['is_apartment']=df['is_apartment'].astype(bool)
print(df.info())
```

```python
#посмотрим на срез данных,видим,что есть аномальные значения почти по всем колонкам
print(df.describe())
```

```python
#Заменим пропущенные значения,аномальные значения( < 2,5 м > 4 м) в колонке ceiling_height на медиану
ceiling_height_median=df['ceiling_height'].median()
df['ceiling_height']=np.where(df['ceiling_height']<2.5,ceiling_height_median,df['ceiling_height'])
df['ceiling_height']=np.where(df['ceiling_height'] > 4.0 ,ceiling_height_median,df['ceiling_height'])
df['ceiling_height']=df['ceiling_height'].fillna(ceiling_height_median)
```

```python
#замена пустых значений в колонках

df.fillna({ 'ceiling_height': df['ceiling_height'].median(),
            'floors_total': 1,
            'balcony': 0,
            'airports_nearest' : df['airports_nearest'].max(),
            'cityCenters_nearest': df['cityCenters_nearest'].max(),
            'parks_around3000': 0,
            'parks_nearest' : df['parks_nearest'].max(),
            'ponds_around3000': 0,
            'ponds_nearest' : df['ponds_nearest'].max()},
          inplace=True)
```


Высоту потолков заполним медианой

Количество этажей заменим на 1 (предположим частный дом)

Балконы меняем на 0 (т.е. пропуск рассматриваем как отсутствие балкона)

Расстояние до центра, парков, водоёмов и аэропортов заменим на максимальные значения.

Соответственно в радиусе 3км данных объектов скорее всего нет, заполняем нулями.

```python
#удаление аномалий в колонке rooms
Q1=df['rooms'].quantile(0.25)
Q3=df['rooms'].quantile(0.75)
IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
df_cleaned= df[(df['rooms']>=lower_bound)&(df['rooms']<=upper_bound)]
```

Аномалии в стоимости,жилой и общей площади,в общем кол-ве этажей я не удаляла,т.к.есть действительно очень высокие здания,например,в Москве(если рассматривать Россию).Как и жилая площадь,может достигать огромных значений(2000 кв.м. в Сочи,например-данные из Google),отсюда и стоимость может быть очень высокой.Плюс это могут быть какие-то люксовые квартиры с авторским дизайном,дорогими материалами и т.п.

```python
#Замена пустых значений в колонке living_area через группировку по комнатам и метода transform
median_values=df_cleaned.groupby('rooms')['living_area'].median()
df_cleaned['living_area']=df_cleaned.groupby('rooms')['living_area'].transform(lambda x: x.fillna(median_values[x.name]))

#Замена пустых значений в колонке kitchen_area на медиану,используя метод группировки по комнатам

df_cleaned.kitchen_area = df_cleaned.groupby('rooms').kitchen_area.transform(lambda x: x.fillna(x.median()))
print(df_cleaned.isnull().sum())
df_cleaned=df_cleaned.dropna(subset='kitchen_area') #удалим оставшиеся  пустые значения
```

```python
print(df_cleaned.info())
```

```python
#4 метод математической статистики для обработки данных
median_last_price=df_cleaned['last_price'].median() #найдем срединное значение цены на квартиры
print(f'Медиана стоимости квартир:{median_last_price}')
mean_ceiling_height=df_cleaned['ceiling_height'].mean() #найдем среднюю высоту потолков
print(f'Средняя высота потолков:{mean_ceiling_height}')
mode_rooms=df_cleaned['rooms'].mode() #узнаем,каких квартир больше всего продается(со скольким кол-вом комнат)
print(f'Мода по количеству комнат:{mode_rooms}')
print(df_cleaned.describe())
```

```python
#тепловая карта корреляции
plt.figure(figsize=(15,10))
corr_matrix=df_cleaned[['last_price','total_area','rooms','ceiling_height','floors_total','living_area','floor','is_apartment','studio','open_plan','kitchen_area','balcony','airports_nearest','cityCenters_nearest','parks_around3000','parks_nearest','ponds_around3000','ponds_nearest']].corr()
sb.heatmap(corr_matrix,annot=True,cmap='coolwarm')
plt.show()
```

```python
pearson_corr,_=st.pearsonr(df_cleaned['cityCenters_nearest'],df_cleaned['last_price'])
spearman_corr,_=st.spearmanr(df_cleaned['cityCenters_nearest'],df_cleaned['last_price'])
print(f'Корреляция Пирсона:{round(pearson_corr,2)}')
print(f'Корреляция Спирмена:{round(spearman_corr,2)}')
```

в данном датасете мы видим отрицательную(обратную) корелляцию относительно стоимости недвижимости и удаленности от центра города,чем ближе к центру города,тем выше стоимость и наоборот.

```python
pearson_corr,_=st.pearsonr(df_cleaned['last_price'],df_cleaned['total_area'])
spearman_corr,_=st.spearmanr(df_cleaned['last_price'],df_cleaned['total_area'])
print(f'Корреляция Пирсона:{round(pearson_corr,2)}')
print(f'Корреляция Спирмена:{round(spearman_corr,2)}')
```

наблюдаем сильную положительную корреляцию по Спирмену относительно стоимости и общей площадью.Чем больше площадь,тем выше стоимость.

```python
#7 Аналитическая задача: Выяснить,какие из признаков оказывают большое влияние на ценообразование недвижимости

#сделаем тест Андерсона на нормальность распределения

result=st.anderson(df['last_price'])
print(f'T-Stat:{result.statistic}')
print(f'Critical-Value:{result.critical_values}')
print(f'Уровень значимости:{result.significance_level}')
for i in range (len(result.critical_values)):
  sl,cv=result.significance_level[i],result.critical_values[i]
  if result.statistic < cv :
    print(f'На уровне значимости {sl}% данные кажутся нормально распределенными')
  else:
    print(f'На уровне значимости {sl}% данные  не кажутся нормально распределенными')
```

```python
# Выдвигаем 1ю гипотезу: Стоимость недвижимости зависит от их общей площади

mean_total_area=df_cleaned['total_area'].mean()
mean_last_price=df_cleaned['last_price'].mean()
high_total_area=df_cleaned[df_cleaned['total_area'] > mean_total_area]['last_price']
low_total_area=df_cleaned[df_cleaned['total_area'] <= mean_total_area]['last_price']
t_stat,p_value = st.mannwhitneyu(high_total_area,low_total_area)
print(f'Т-statistic: {t_stat}')
print(f'P_value: {p_value}')
if p_value < 0.05:
  print('Нулевая гипотеза (H0) принимается,есть статистически значимые различия.Стоимость недвижимости зависит от общей площади.')
else:
  print('Нулевая гипотеза (Н0) отклоняется,принимается альтернативная (Н1): стоимость недвижимости не зависит от общей площади.Нет статистически значимых различий')
```

Применяем тест Манна-Yитни т.к.данные распределены ненормально

```python
# Выдвинем 2ю гипотезу: Стоимость недвижимости зависит от их удаленности от центра города
mean_cityCenters=df_cleaned['cityCenters_nearest'].mean()
mean_last_price=df_cleaned['last_price'].mean()
high_cityCenters=df_cleaned[df_cleaned['cityCenters_nearest'] > mean_cityCenters]['last_price']
low_cityCenters=df_cleaned[df_cleaned['total_area'] <= mean_cityCenters]['last_price']
t_stat,p_value = st.mannwhitneyu(high_cityCenters,low_cityCenters)
print(f'Т-statistic: {t_stat}')
print(f'P_value: {p_value}')
if p_value < 0.05:
  print('Нулевая гипотеза (H0) принимается,есть статистически значимые различия.Стоимость недвижимости зависит от их удалённости от центра города.')
else:
  print('Нулевая гипотеза (Н0) отклоняется,принимается альтернативная (Н1): стоимость недвижимости не зависит от их удалённости от центра города.Нет статистически значимых различий')
```

```python
# Выдвинем 3ю гипотезу: Стоимость недвижимости не зависит от кол-ва комнат

mean_room=df_cleaned['rooms'].mean()
mean_last_price=df_cleaned['last_price'].mean()
more_room=df_cleaned[df_cleaned['rooms'] > mean_room]['last_price']
less_room=df_cleaned[df_cleaned['rooms'] <= mean_room]['last_price']
t_stat,p_value = st.mannwhitneyu(more_room,less_room)
print(f'Т-statistic: {t_stat}')
print(f'P_value: {p_value}')
if p_value > 0.05:
  print('Нулевая гипотеза (H0) принимается, нет статистически значимых различий.Стоимость недвижимости не зависит от количества комнат.')
else:
  print('Нулевая гипотеза (Н0) отклоняется,принимается альтернативная (Н1): стоимость недвижимости зависит от количества комнат.Есть статистически значимые различия')
```

из анализа данных видно,что на стоимость недвижимости влияет в большей степени их общая  и жилая площадь,удаленность от центра города.Кол-во комнат,высота потолков также могут повлиять на стоимость.
